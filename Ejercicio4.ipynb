```python
import tkinter as tk


class Programador:
    def __init__(self, nombre, apellidos):
        self.nombre = nombre
        self.apellidos = apellidos


class EquipoMaratonProgramacion:
    def __init__(self, nombre_equipo, universidad, lenguaje):
        self.nombre_equipo = nombre_equipo
        self.universidad = universidad
        self.lenguaje = lenguaje
        self.programadores = []

    def esta_lleno(self):
        return len(self.programadores) >= 3

    def anadir(self, programador):
        if self.esta_lleno():
            raise Exception("El equipo está completo.")
        self.programadores.append(programador)

    @staticmethod
    def validar_campo(campo):
        for c in campo:
            if c.isdigit():
                raise Exception("El nombre o apellido no puede contener números.")

        if len(campo) > 20:
            raise Exception("La longitud no debe ser superior a 20 caracteres.")


class Ventana:
    def __init__(self, root):
        self.root = root
        self.root.title("Maratón de Programación")
        self.root.geometry("700x600")

        self.equipo = None

        tk.Label(root, text="Nombre del equipo").pack()
        self.txtEquipo = tk.Entry(root, width=40)
        self.txtEquipo.pack()

        tk.Label(root, text="Universidad").pack()
        self.txtUniversidad = tk.Entry(root, width=40)
        self.txtUniversidad.pack()

        tk.Label(root, text="Lenguaje de programación").pack()
        self.txtLenguaje = tk.Entry(root, width=40)
        self.txtLenguaje.pack()

        tk.Button(
            root,
            text="Crear Equipo",
            command=self.crear_equipo
        ).pack(pady=10)

        tk.Label(root, text="Nombre del integrante").pack()
        self.txtNombre = tk.Entry(root, width=40)
        self.txtNombre.pack()

        tk.Label(root, text="Apellidos del integrante").pack()
        self.txtApellidos = tk.Entry(root, width=40)
        self.txtApellidos.pack()

        tk.Button(
            root,
            text="Agregar Integrante",
            command=self.agregar_integrante
        ).pack(pady=10)

        tk.Label(root, text="Resultados").pack()

        self.area = tk.Text(root, width=80, height=20)
        self.area.pack()

    def crear_equipo(self):
        nombre = self.txtEquipo.get()
        universidad = self.txtUniversidad.get()
        lenguaje = self.txtLenguaje.get()

        self.equipo = EquipoMaratonProgramacion(
            nombre,
            universidad,
            lenguaje
        )

        self.area.delete("1.0", tk.END)

        self.area.insert(
            tk.END,
            f"Nombre del equipo = {nombre}\n"
        )

        self.area.insert(
            tk.END,
            f"Universidad = {universidad}\n"
        )

        self.area.insert(
            tk.END,
            f"Lenguaje de programación = {lenguaje}\n\n"
        )

        self.area.insert(
            tk.END,
            "Datos de los integrantes del equipo\n\n"
        )

    def agregar_integrante(self):
        try:
            if self.equipo is None:
                raise Exception(
                    "Primero debe crear el equipo."
                )

            nombre = self.txtNombre.get()
            apellidos = self.txtApellidos.get()

            EquipoMaratonProgramacion.validar_campo(nombre)
            EquipoMaratonProgramacion.validar_campo(apellidos)

            programador = Programador(
                nombre,
                apellidos
            )

            self.equipo.anadir(programador)

            self.area.insert(
                tk.END,
                f"Nombre del integrante = {nombre}\n"
            )

            self.area.insert(
                tk.END,
                f"Apellidos del integrante = {apellidos}\n\n"
            )

            self.txtNombre.delete(0, tk.END)
            self.txtApellidos.delete(0, tk.END)

        except Exception as e:
            self.area.insert(
                tk.END,
                f"ERROR: {e}\n\n"
            )


root = tk.Tk()
app = Ventana(root)
root.mainloop()
```

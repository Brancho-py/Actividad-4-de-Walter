```python
import tkinter as tk
import math

def calcular():

    try:
        valor = float(campo_valor.get())


        if valor < 0:
            etiqueta_log.config(
                text="El valor debe ser un número positivo para calcular el logaritmo"
            )
        else:
            etiqueta_log.config(
                text=f"Logaritmo neperiano = {math.log(valor):.2f}"
            )


        if valor < 0:
            etiqueta_raiz.config(
                text="El valor debe ser un número positivo para calcular la raíz cuadrada"
            )
        else:
            etiqueta_raiz.config(
                text=f"Raíz cuadrada = {math.sqrt(valor):.2f}"
            )

    except ValueError:
        etiqueta_log.config(
            text="El valor debe ser numérico para calcular el logaritmo"
        )
        etiqueta_raiz.config(
            text="El valor debe ser numérico para calcular la raíz cuadrada"
        )

def limpiar():
    campo_valor.delete(0, tk.END)
    etiqueta_log.config(text="Logaritmo neperiano =")
    etiqueta_raiz.config(text="Raíz cuadrada =")


ventana = tk.Tk()
ventana.title("Cálculos Numéricos")
ventana.geometry("400x280")

tk.Label(ventana, text="Valor numérico:").pack(pady=10)

campo_valor = tk.Entry(ventana)
campo_valor.pack()

tk.Button(ventana, text="Calcular", command=calcular).pack(pady=10)
tk.Button(ventana, text="Limpiar", command=limpiar).pack()

etiqueta_log = tk.Label(ventana, text="Logaritmo neperiano =")
etiqueta_log.pack(pady=15)

etiqueta_raiz = tk.Label(ventana, text="Raíz cuadrada =")
etiqueta_raiz.pack()

ventana.mainloop()
```
